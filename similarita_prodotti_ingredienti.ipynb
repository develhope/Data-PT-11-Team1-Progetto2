```python
import pandas as pd
import csv

SIM_THRESHOLD = 0.1   # soglia minima di similarit
OUTPUT_FILE = "ingredient_similarity.csv"

df = pd.read_excel("ingr_cat.xlsx")

df = df[df["is_functional"] == 1]

df = df.dropna(subset=["code", "ingredient_id"])


product_ingredients = (
    df.groupby("code")["ingredient_id"]
    .apply(lambda x: set(x))
    .to_dict()
)

products = list(product_ingredients.keys())

print(f"Prodotti caricati: {len(products)}")


def jaccard(set_a, set_b):
    if not set_a or not set_b:
        return 0.0

    intersection = set_a.intersection(set_b)
    union = set_a.union(set_b)

    return len(intersection) / len(union)


with open(OUTPUT_FILE, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["product_1", "product_2", "ingredient_similarity"])

    for i in range(len(products)):
        prod_a = products[i]
        ingr_a = product_ingredients[prod_a]

        for j in range(i + 1, len(products)):
            prod_b = products[j]
            ingr_b = product_ingredients[prod_b]

            sim = jaccard(ingr_a, ingr_b)

            if sim >= SIM_THRESHOLD:
                writer.writerow([
                    prod_a,
                    prod_b,
                    round(sim, 4)
                ])


        if i % 50 == 0:
            print(f"Processati {i}/{len(products)} prodotti")

```
