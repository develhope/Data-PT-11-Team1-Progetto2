```python
import pandas as pd

# Dizionario macro-categorie (beauty/skincare dal tuo dataset)
macro_rules = {
    "Hair_Care": ["hair", "shampoo", "conditioner", "shampoos", "hair care", "styling", "hair mask"],
    "Body_Care": ["body", "shower", "gel douche", "body cream", "body lotion", "body wash"],
    "Face_Care": ["face", "facial", "visage", "face cream", "moisturizer", "cleanser"],
    "Hygiene": ["hygiene", "soap", "savon", "toothpaste", "dentifrice", "mouthwash"],
    "Deodorants": ["deodorant", "deodorants", "antiperspirant"],
    "Sun_Care": ["sun", "solaire", "sunscreen", "uv", "spf"],
    "Makeup": ["makeup", "maquillage", "foundation", "lip", "eye", "mascara"],
    "Shaving": ["shaving", "rasage", "aftershave", "razor"],
    "Nails": ["nail", "ongle", "nail polish", "vernis"],
    "Lip_Care": ["lip", "lèvres", "lip balm", "baume lèvres"],
    "Eye_Care": ["eye", "yeux", "eye cream", "eye shadow"],
    "Hand_Care": ["hand", "mains", "hand cream"],
    "Foot_Care": ["foot", "pieds", "foot cream"],
    "Baby_Care": ["baby", "bébé", "kids"],
    "Men_Care": ["men", "homme", "beard"],
    "Intimate": ["intimate", "intima", "lubrifiant"],
    "Cleaning": ["cleaning", "nettoyant", "dishwashing"],
    "Medicinal": ["medicine", "médicament", "antiseptic"],
    "Perfumes": ["perfume", "parfum", "eau de"],
    "Masks": ["mask", "masque"],
    "Oils": ["oil", "huile", "huiles essentielles"]

}

def assign_macro(text):
    if pd.isna(text):
        return "Other"
    text_lower = str(text).lower()
    for macro, keywords in macro_rules.items():
        if any(kw in text_lower for kw in keywords):
            return macro
    return "Other"

# 1. Assegna macro-categorie
cat["macro_category"] = cat.apply(assign_macro)

# 2. Crea ID numerici (1,2,3...)
unique_macros = sorted(cat["macro_category"].unique())
macro_id_map = {macro: i+1 for i, macro in enumerate(unique_macros)}
cat["macro_id"] = cat["macro_category"].map(macro_id_map)

# 3. Tabella mappatura
mappatura = pd.DataFrame({
    "macro_id": macro_id_map.values(),
    "macro_category": macro_id_map.keys()
})

# 4. Statistiche

print(mappatura)

```
